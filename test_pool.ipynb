```python
import torch
import torch.nn as nn
import torch.nn.functional as F

from converted_KAN.relumaxpool2d import ReLUMaxPool2dKAN
from converted_KAN.avgpool2dkan import AvgPool2dKAN


if __name__ == "__main__":
    # MaxPool via ReLU reduction
    pool_ref = nn.MaxPool2d(kernel_size=2, stride=2)
    pool_relu_kan = ReLUMaxPool2dKAN(kernel_size=2, stride=2)

    x = torch.randn(3, 4, 8, 8)

    y_ref = pool_ref(x)
    y_relu = pool_relu_kan(x)

    print("Max abs diff (max pool):", (y_ref - y_relu).abs().max().item())

    # AvgPool via unfold + sum / k
    pool_avg_ref = nn.AvgPool2d(kernel_size=2, stride=2)
    pool_avg_kan = AvgPool2dKAN(kernel_size=2, stride=2)

    y_avg_ref = pool_avg_ref(x)
    y_avg = pool_avg_kan(x)

    print("Max abs diff (avg pool):", (y_avg_ref - y_avg).abs().max().item())
```
